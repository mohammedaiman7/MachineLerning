```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
cd = pd.read_csv("canada_per_capita_income.csv")
cd
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per capita Income(US$)")
plt.scatter(cd.year,cd.per_capita_income,color="red",)
```

```python
reg=linear_model.LinearRegression()
reg.fit(cd[['year']],cd.per_capita_income)
```

```python
reg.predict([[2017]])
```
